```python
from transformers import pipeline
```

```python
text = "Python is an excellent programming language known for its simplicity, readability, and versatility. It is widely used in web development, data science, artificial intelligence, machine learning, automation, and backend engineering. With a vast ecosystem of libraries and frameworks, Python enables developers to build scalable, efficient, and high-performance applications across a wide range of domains."
pipe= pipeline("text-to-speech", model= "suno/bark-small")
output = pipe(text)
```

```python
output
```

```python
from IPython.display import Audio
Audio(output["audio"], rate= output['sampling_rate'])
```
